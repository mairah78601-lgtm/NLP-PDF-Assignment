```python
# PDF Reading
import PyPDF2

pdf_file = open("sample.pdf", "rb")

pdf_reader = PyPDF2.PdfReader(pdf_file)

total_pages = len(pdf_reader.pages)

print("Total Pages:", total_pages)

full_text = ""

for page in pdf_reader.pages:
    full_text += page.extract_text()

print(full_text[:1000])
```

```python
# Import NLP Libraries
import re
import string
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Lowercase
text = full_text.lower()

print(text[:500])
```

```python
# Remove Numbers Using Regex
text = re.sub(r'\d+', '', text)

print(text[:500])
```

```python
# Remove Special Symbols
text = re.sub(r'[^a-zA-Z\s]', '', text)

print(text[:500])
```

```python
# Remove Extra Spaces
text = re.sub(r'\s+', ' ', text).strip()

print(text[:500])
```

```python
# Remove Punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

print(text[:500])
```

```python
# Tokenization
tokens = word_tokenize(text)

print(tokens[:50])
```

```python
# Stop Word Removal
stop_words = set(stopwords.words('english'))

stop_word_count = 0
valid_words = []

for word in tokens:
    if word in stop_words:
        stop_word_count += 1
    else:
        valid_words.append(word)

print("Stop Words Count:", stop_word_count)
print("Valid Words Count:", len(valid_words))
```

```python
# Stemming
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in valid_words]

print(stemmed_words[:50])
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in valid_words]

print(lemmatized_words[:50])
```

```python
# One Hot Encoding
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

sample_words = valid_words[:20]

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(
    pd.DataFrame(sample_words, columns=['Words'])
)

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(['Words'])
)

print(encoded_df)
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

cleaned_text = " ".join(valid_words)

vectorizer = TfidfVectorizer(max_features=20)

tfidf_matrix = vectorizer.fit_transform([cleaned_text])

feature_names = vectorizer.get_feature_names_out()

tfidf_values = tfidf_matrix.toarray()

tfidf_df = pd.DataFrame(
    tfidf_values,
    columns=feature_names
)

print(tfidf_df)
```

```python
# Plotly Scatter Plot
import plotly.express as px

words = feature_names
scores = tfidf_values[0]

plot_df = pd.DataFrame({
    'Words': words,
    'TF-IDF Score': scores
})

fig = px.scatter(
    plot_df,
    x='Words',
    y='TF-IDF Score',
    title='TF-IDF Scatter Plot'
)

fig.update_layout(
    xaxis_title='Words',
    yaxis_title='TF-IDF Scores'
)

fig.show()
```
